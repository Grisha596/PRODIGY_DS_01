```python
#Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading data for total, male, and female populations
total_population_url = "C:\\Users\\Acer\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_6011311.csv"
male_population_url = "C:\\Users\\Acer\\Downloads\\API_SP.POP.TOTL.MA.IN_DS2_en_csv_v2_5996754.csv"
female_population_url = "C:\\Users\\Acer\\Downloads\\API_SP.POP.TOTL.FE.IN_DS2_en_csv_v2_5996755.csv"
```

```python
df_total = pd.read_csv(total_population_url, skiprows=4, usecols=lambda x: x not in ['Unnamed: 67'])
df_male = pd.read_csv(male_population_url, skiprows=4, usecols=lambda x: x not in ['Unnamed: 67'])
df_female = pd.read_csv(female_population_url, skiprows=4, usecols=lambda x: x not in ['Unnamed: 67'])
```

```python
#Choosing a country or region to visualize
country_to_visualize = 'World'
```

```python
#Filtering data for the selected country or region
selected_country_data_total = df_total[df_total['Country Name'] == country_to_visualize]
selected_country_data_male = df_male[df_male['Country Name'] == country_to_visualize]
selected_country_data_female = df_female[df_female['Country Name'] == country_to_visualize]
```

```python
#Plotting a bar chart for the distribution of male and female populations
years = selected_country_data_total.columns[4:]
plt.figure(figsize=(12, 6))
plt.bar(years, selected_country_data_male.iloc[0, 4:], label='Male', color='#008080', alpha=0.7)
plt.bar(years, selected_country_data_female.iloc[0, 4:], label='Female', bottom=selected_country_data_male.iloc[0, 4:], color='#FF1493', alpha=0.7)

plt.title(f'Population Distribution in {country_to_visualize} by Gender (1960-2022)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
